```python
#
#COVARIANCE wrt SP500
```

```python
# Calculate covariance of GBTC wrt SP500

GBTC_cov_sp = combined_df["GBTC_return"].cov(combined_df["SP500 return"])
```

```python
# Computing GBTC beta with respect to NASDAQ
GBTC_beta_nq = GBTC_cov_nq / variance_NASDAQ

#display NASDAQ variance and beta value
print(f" NASDAQ variance: {variance_NASDAQ}, and GBTC beta value: {GBTC_beta_nq}")
```

```python
# Computing ARKK beta with respect to NASDAQ
ARKK_beta_nq = ARKK_cov_nq / variance_NASDAQ

#display SP500 variance and beta value
print(f" Nasdaq variance: {variance_NASDAQ}, and ARKK beta value: {ARKK_beta_nq}")
```

```python
# Calculate 30-day rolling covariance of ETFs vs. NASDAQ and plot the data
rolling_GBTC_covariance = combined_df["GBTC_return"].rolling(window=30).cov(combined_df["NASDAQ return"])
rolling_ARKK_covariance = combined_df["ARKK return"].rolling(window=30).cov(combined_df["NASDAQ return"])
```

```python
# Calculate 30-day rolling covariance of ETFs vs. SP500 and plot the data 
rolling_GBTC_covariance1 = combined_df["GBTC_return"].rolling(window=30).cov(combined_df["SP500 return"])
rolling_ARKK_covariance1 = combined_df["ARKK return"].rolling(window=30).cov(combined_df["SP500 return"])
```

```python
#
```

```python
# Calculate 30-day rolling variance of NASDAQ
rolling_variance_NDQ = combined_df["NASDAQ return"].rolling(window=30).var()
```

```python
# Calculate 30-day rolling variance of SP500 
rolling_variance_SP = combined_df["SP500 return"].rolling(window=30).var()
```

```python
#
```

```python
# Calculate 30-day rolling beta of ETFs and plot the data wrt NASDAQ
rolling_GBTC_beta = rolling_GBTC_covariance / rolling_variance_NDQ
rolling_ARKK_beta = rolling_ARKK_covariance / rolling_variance_NDQ
```

```python
# Calculate 30-day rolling beta of ETFs and plot the data wrt sp500
rolling_GBTC_beta1 = rolling_GBTC_covariance1 / rolling_variance_SP
rolling_ARKK_beta1 = rolling_ARKK_covariance1 / rolling_variance_SP
```

```python
#
```

```python
# Set the figure and plot the different ETFs beta values as multiple trends on the same figure
ax = rolling_GBTC_beta.plot(figsize=(20, 10), title="Rolling 30-Day Beta w.r.t. NASDAQ of GBTC ETF and ARKK ETF")
rolling_ARKK_beta.plot(ax=ax)


# Set the legend of the figure
ax.legend(["GBTC", "ARKK"])
```

```python
# Set the figure and plot the different ETFs beta values as multiple trends on the same figure
ax = rolling_GBTC_beta1.plot(figsize=(20, 10), title="Rolling 30-Day Beta w.r.t. S&P500 of GBTC ETF and ARKK ETF")
rolling_ARKK_beta1.plot(ax=ax)


# Set the legend of the figure
ax.legend(["GBTC", "ARKK"])
```
